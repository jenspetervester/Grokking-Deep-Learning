_multiple inputs: updating weights is done by scaling proportionally with the input._
e.g. multiply the weight_delta with the previous inputs, then for each weight, subtract the weight_delta multiplied with the alpha value

```python
def neural_network(input, weights):
    res = 0
    for i in range(len(weights)):
        res += input[i] * weights[i]
    return res
```

```python
def ele_mul(scalar, vector):
    res = [0,0,0]
    for i in range(len(res)):
        res[i] = vector[i] * scalar
    return res
```

```python
#input data
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1,1,0,1]
true = win_or_lose_binary[0]

alpha = 0.003

weights = [0.1, 0.2, -0.1]

input = [toes[0], wlrec[0], nfans[0]]
```

* freezing the first weight to explore what happens

```python
for it in range(300):
    pred = neural_network(input, weights)
    error = (pred - true) ** 2
    delta = pred - true
    weight_deltas = ele_mul(delta, input)
    #freezing
    #weight_deltas[0]= 0
    
    print("Iteration:" + str(it))
    print( "Pred:" + str(pred))
    print("Error:" + str(error))
    print("Delta:" + str(delta))
    print("Weights:" + str(weights))
    print("Weight_Deltas:")
    print(str(weight_deltas))
    print()
    
    for i in range(len(weights)):
        weights[i] -= alpha * weight_deltas[i]    
```

```python
def neural_network(inputs, weights):
    pred = vops.vect_mat_mul(input, weights)
    return pred
```

```python
alpha = 0.01
input = [toes[0],wlrec[0],nfans[0]]
#input = [toes[0],wlrec[0],nfans[0], blab[0]] # expand to work with one more input
true = [hurt[0], win[0], sad[0]]

pred = neural_network(input, weights)

error = [0,0,0]
delta = [0,0,0]

for i in range(len(true)):
    error[i] = (pred[i] - true[i]) ** 2
    delta[i] = pred[i] - true[i]

print("error: {}, delta: {}".format(error, delta))

weight_deltas = vops.outer_prod(input, delta)

print("weight_deltas: {}".format(weight_deltas))

#update weights
for i in range(len(weights)):
    for j in range(len(weights[0])):
        weights[i][j] -= alpha * weight_deltas[i][j]
```
