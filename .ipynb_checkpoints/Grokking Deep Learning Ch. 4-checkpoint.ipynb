```python
weight = 0.1
lr = 0.01 # "learning step size?"

def neural_network(input, weight):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1] # won

input = number_of_toes[0]
true = win_or_lose_binary[0]

pred = neural_network(input, weight)

error = (pred - true) ** 2
print("error: ", error)

p_up = neural_network(input, weight+lr)
e_up = (p_up - true) ** 2

print("error up:", e_up)

p_dn = neural_network(input, weight-lr)
e_dn = (p_dn - true) ** 2

print("error down:", e_dn)

if (error > e_dn or error > e_up):
    if (e_dn < e_up):
        weight -= lr
    if (e_dn > e_up):
        weight += lr

print("updated weight: ", weight)
```

Hot and Cold Learning

```python
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001
for iteration in range(1101):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2
    print("Error:" + str(error) + " Prediction: " + str(prediction))
    up_prediction = input * (weight + step_amount)
    up_error = (up_prediction - goal_prediction) ** 2
    down_prediction = input * (weight - step_amount)
    down_error = (down_prediction - goal_prediction) ** 2
    if (down_error < up_error):
        weight -= step_amount
    if (down_error > up_error):
        weight += step_amount
print("weight: " + str(weight))
```

```python
weight = 0.5
input = 0.5
goal_prediction = 0.8
alpha = 0.01 # control how fast the network learns to avoid overshooting

for i in range(3000):
    pred = input * weight
    error = (pred - goal_prediction) ** 2
    direction_and_amount = (pred - goal_prediction) * input
    weight = weight - direction_and_amount * alpha
    print ("Error: " + str(error) + " Prediction: " + str(pred))
```

# Simple Gradient Descent

_core_: ** "error = ((input * weight) - goal_prediction) ** 2"
_manipulate the weight until error is 0. Exact relationship.

Calculus: "All you really need to know is what the derivative represents. It's the relationship between two variables in a function so that you can know how much one changes when you change the other"
"It's just the sensitivity between two variables"
"The derivative defines the realtionship between each weight and how much we missed. In other words, it tells how much changing a weight contributed to the error"

```python
weight = 0.0
input = 1.1
goal_prediction = 0.8

for i in range(4):
    pred = input * weight
    error = (pred - goal_prediction) ** 2
    delta = pred - goal_prediction
    weight_delta = delta * input
    print("Error: " + str(error) + " weight delta: " + str(weight_delta) + " prediction: " + str(pred))
    weight = weight - weight_delta

print("Error: " + str(error) + " weight delta: " + str(weight_delta) + "prediction: " + str(pred))
```

```python
weight = 0.5
input = 2
goal_prediction = 0.8
# try without alpha: delta grows way too fast
alpha = 0.1 # prevent overshooting, just guessing. Indicator: error goes up proportional to number of iterations

# "finding the appropriate alpha, even for state-of-theart
# neural networks, is often done simply by guessing"

for i in range(20):
    pred = input * weight
    error = (pred - goal_prediction) ** 2
    #delta = pred - goal_prediction
    #weight_delta = input * delta
    #weight = weight - weight_delta * alpha
    derivative = input * (pred - goal_prediction)
    weight = weight - (alpha * derivative)
   # print("Error: " + str(error) + " derivative: " + str(derivative) + " prediction: " + str(pred))
    print("Error: {:12.10f}, derivative: {:10.6f}, prediction: {:10.5f}".format(error, derivative, pred))
    
print("weight: {}".format(weight))
```

```python
import matplotlib
import matplotlib.pyplot as plt

def calc_slope(input, weight, goal_prediction):
    return 2 * input * (-goal_prediction + input * weight)

def plot_slope(input, weight, goal_prediction):
    slope = calc_slope(input, weight, goal_prediction)
    error = ((input * weight) - goal_prediction) ** 2
    x0 = weight
    y0 = error
    x1 = weight + 0.2
    #point slope formula: y - y1 = m(x-x1) ==> y = m(x-x1) + y1
    y1 = slope*(x1 - weight) + err0
    x2 = weight - 0.2
    y2 = slope*(x2 - weight) + err0 
    plt.plot([x2, x0, x1])
    
```
